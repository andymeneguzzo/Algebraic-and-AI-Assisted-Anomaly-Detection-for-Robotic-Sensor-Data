# Algebraic and AI-Assisted Anomaly Detection for Robotic Sensor Data

## Model Comparison

- Load all saved anomaly scores from the different methods implemented
- Merge and align per dataset
- Perform data cleaning
- Compute correlations, stats and ranking
- Plot cross-method patterns
- Export summary to feed Gemini 2.0flash for AI-assisted analysis

```python
# --- 1. Imports ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import json

SEED = 42
np.random.seed(SEED)
plt.style.use("seaborn-v0_8-darkgrid")
pd.set_option("display.max_columns", None)

# --- 2. Paths ---
results_path = Path("../results")
files = sorted(results_path.glob("*.csv"))

print("[INFO] Result files found:")
for f in files:
    print(f" - {f.name}")

# --- 3. Load all score files ---
datasets = {}
for f in files:
    name = f.stem.split("_")[0]  # e.g. "lp1"
    df = pd.read_csv(f)
    if name not in datasets:
        datasets[name] = df
    else:
        # merge different method outputs for same dataset
        datasets[name] = pd.concat([datasets[name], df], axis=1)

print(f"[INFO] Loaded {len(datasets)} datasets with all methods merged.")

# Inspect one dataset
for k,v in datasets.items():
    print(f"\n{k}: {v.shape} columns: {list(v.columns)}")
    display(v.head(3))
    break


"""Data cleaning"""
# remove duplicate cols and clean nans (if any)
for name, df in datasets.items():
    df = df.loc[:,~df.columns.duplicated()]
    df = df.dropna(axis=0, how="any")
    datasets[name] = df

# collect list of method cols
method_columns = sorted(set(sum([list(df.columns) for df in datasets.values()], [])))
print(f"[INFO] Total unique score cols: {method_columns}")


"""Correlation analysis"""
for name, df in datasets.items():
    print(f"\n=== Dataset: {name} ===")
    corr = df.corr()
    plt.figure(figsize=(7,6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
    plt.title(f"{name} — Cross-Method Correlation Matrix")
    plt.tight_layout()
    plt.show()


"""Summary stats and rankings"""
summary = []
for name, df in datasets.items():
    stats = df.describe().T
    stats["Dataset"] = name
    summary.append(stats)
summary_df = pd.concat(summary)
display(summary_df)

# Compute per-dataset mean/std per method
mean_std = summary_df[["Dataset", "mean", "std"]].reset_index().rename(columns={"index": "Method"})
pivot_mean = mean_std.pivot(index="Dataset", columns="Method", values="mean")
pivot_std = mean_std.pivot(index="Dataset", columns="Method", values="std")

plt.figure(figsize=(12,5))
sns.heatmap(pivot_mean, cmap="crest")
plt.title("Average Anomaly Scores per Method")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,5))
sns.heatmap(pivot_std, cmap="mako")
plt.title("Std. Deviation of Anomaly Scores per Method")
plt.tight_layout()
plt.show()


"""Score Normalization and Ranking"""
def normalize(s):
    return (s-s.min()) / (s.max()-s.min()+1e-8)

rank_summary = []
for name, df in datasets.items():
    norm_df = df.apply(normalize)
    mean_scores = norm_df.mean().sort_values()
    rank_summary.append(mean_scores.rename(name))
rank_df = pd.DataFrame(rank_summary).T
display(rank_df)

plt.figure(figsize=(8,4))
sns.heatmap(rank_df, annot=True, cmap="viridis", fmt=".2f")
plt.title("Mean normalized anomaly scores (lower ≈ more sensitive)")
plt.tight_layout()
plt.show()


"""Cross-dataset method corr"""
method_means = []
for name, df in datasets.items():
    method_means.append(df.mean().rename(name))
method_means_df = pd.DataFrame(method_means)
corr_methods = method_means_df.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr_methods, annot=True, cmap="cividis", fmt=".2f")
plt.title("Cross-Dataset Correlation between Methods")
plt.tight_layout()
plt.show()

print("\n[INFO] Method-level correlation across all datasets:")
display(corr_methods.round(3))


"""JSON for Gemini 2.0"""
gemini_summary = {}

for name, df in datasets.items():
    summary_info = {
        "method_statistics": df.describe().T.to_dict(),
        "method_correlations": df.corr().to_dict(),
        "mean_scores": df.mean().to_dict(),
        "std_scores": df.std().to_dict(),
    }
    gemini_summary[name] = summary_info

# Save JSON summary for Gemini analysis
json_path = results_path / "gemini_summary.json"
with open(json_path, "w") as f:
    json.dump(gemini_summary, f, indent=2)
print(f"[INFO] Saved Gemini analysis summary → {json_path}")
```
